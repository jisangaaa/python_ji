```python
try:
    f = open('뜬금파일','r')
except FileNotFoundError as e :
    if e:
        print('알고있었지')
    print(e)
finally :
    print('코드작성을 마칩니다.')
    
```

```python
# 지정한 이름의 예외처리, as 로 별칭가능.
try:
    4/0
except ZeroDivisionError as e:
    print('0으로 나눌 수 없어요',e)
```

```python
# try-except-finally 3단계도 가능, finally는 에러가나도 무조건 처리하는 것(여기서는 파일 닫기)
try:
    f = open('foo.txt','w')
    f.write(4/0)
except:
    print("0으로 나누지마세요")
finally:
    f.close()
```

```python
# 테스트 해볼것 2개 이상의 오류를 동일하게 처리
try:
    a =1,2
    print(a[512])
    4/0
except (ZeroDivisionError, IndexError) as e:
    print(e)
```

```python
# try-else 문
try:
    age = int(input("나이를 입력하세요"))
except:
    print("입력이 정확하지 않습니다.")
else:
    if age <= 18:
        print('미성년자는 출입금지')
    else:
        print('환영합니다.')
finally:
    print('='*20+'\n분석끝.')
```

```python
class Add:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def add(self):
        return self.a+self.b
add = Add(3,8)
result = add.add()
print('결과는', result)
```

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def multiply(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Error! Division by zero."
        return self.num1 / self.num2

# 예제 사용법
calc = Calculator(10, 5)
print("덧셈:", calc.add())         # 결과: 15
print("뺄셈:", calc.subtract())    # 결과: 5
print("곱셈:", calc.multiply())    # 결과: 50
print("나눗셈:", calc.divide())     # 결과: 2.0
```

```python
positive_words = ["좋아", "행복", "멋져", "최고", "사랑해"]
negative_words = ["싫어", "화나", "짜증", "별로", "나빠"]

text = input("당신의 기분을 한 문장으로 표현해주세요.")

pnum = 0
text = text.split(" ")
for v in positive_words:
    for s in text:
        if v == s:
            pnum += 1
            print(pnum)
for i in negative_words:
    for j in text:
        if i == j:
            pnum -= 1
            print(pnum)
print(pnum)
```
