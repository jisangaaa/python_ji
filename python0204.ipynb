# 문자열 포매팅

```python
print(' i eat 3 apples.')
print(' i eat %d apples.'%17)
print(' i eat %s apples.'%'five')  # %s 는 문자 숫자 둘다 가능
```

```python
# 코드	#설명
# %s	#문자열(String)
# %c	#문자 1개(character)
# %d	#정수(Integer)
# %f	#부동소수(floating-point)
# %o	#8진수
# %x	#16진수
# %%	#Literal % (문자 % 자체)
```

```python
print('사람 몸의 수분은 60% 입니다.')
print('사람 몸의 수분은 %d%% 입니다.'%65)
```

```python
print('십진수로 123은 123이다.')
print('십진수로 123은 %d이다.'%123)
print('십진수로 %d은 %d이다.'%(123,123),'\n')
print('십진수로 %d은 16진수로는 %x이고 8진수로는 %o 이다.'%(123,123,123))
```

```python
print('%10s' % 'hi' )
```

```python
print('%-10sjane' % 'hi' )
```

```python
print('3.141592''\n') 
print('%d'%3.141592+ '\n')
print('%f'%3.141592 ,'\n')
print('%0.4f'%3.141592 ,'\n')
print('%.4f'%3.141592 ,'\n')
print('%8.4f'%3.141592 ,'\n')
print('%20.6f'%3.141592 ,'\n')
```
